Ao considerar esses fatores, é possível obter uma visão mais abrangente e precisa do bem-estar das pessoas em uma nação.

Nessa apresentação vou discorrer sobre a percepção de felicidade nas características da Liberdade para fazer escolhas e Generosidade, desde 2016 até a o ano de 2022, e mostrar como impactou no resultado do relatório.


```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
```

Inicio apresentando os dados referente ao ano de 2016, com os 5 primeiros países do ranking em azul e os 5 últimos países em vermelho.
Para esse ano, levamos em conta somente a percepção da liberdade para fazer escolhas, e observamos os valores discrepantes da Síria e de Burundi, países onde quase não há liberdade de escolha e os 5 primeiros países.

Dados ano 2016

```python
dados2016 = pd.read_csv('/content/2016.csv')
dados2016
```

Apresentando os dados dos 5 países melhores colocados e os 5 piores colocados com gráfico de barras simples.

```python
# Colunas desejadas
nomes_paises = pd.concat([dados2016['Country'][:5], dados2016['Country'][-5:]])
valores = pd.concat([dados2016['Freedom'][:5], dados2016['Freedom'][-5:]])

# Cores para as barras azul para os 5 primeiros, vermelho para os últimos 5
cores = ['blue'] * 5 + ['red'] * 5

# Criando o gráfico de barras
plt.bar(nomes_paises, valores, color=cores)

# Personalizando o eixo x com os rótulos adequados
plt.xticks(rotation=90)

# Exibir o gráfico
plt.show()
```

Dados ano 2017

```python
dados2017 = pd.read_csv('/content/2017.csv')
dados2017
```

Agora atentem para o ano de 2017, onde analiso Liberdade e Generosidade juntas.
Síria e Burundi, que continuam com a taxa de Liberdade ainda baixa, têm na Generosidade valores bastantes expressivos. Síria chega a ultrapassar os valores dos 5 primeiros países do ranking. Burundi se aproxima em muito da Finlândia.

Apresentando os dados dos 5 países melhores colocados e dos 5 piores com o gráfico de barras agrupado.

```python
# Colunas desejadas
nomes_paises = pd.concat([dados2017['Country'][:5], dados2017['Country'][-5:]])
valores1 = pd.concat([dados2017['Freedom'][:5], dados2017['Freedom'][-5:]])
valores2 = pd.concat([dados2017['Generosity'][:5], dados2017['Generosity'][-5:]])

# Definindo a largura das barras
largura_barra = 0.35

# Definindo a posição das barras no eixo x
posicao_barras1 = np.arange(len(nomes_paises))
posicao_barras2 = posicao_barras1 + largura_barra

# Criando o gráfico de barras agrupado
plt.bar(posicao_barras1, valores1, width=largura_barra, label='Freedom')
plt.bar(posicao_barras2, valores2, width=largura_barra, label='Generosity')

# Personalizando o eixo x com os rótulos adequados
plt.xticks(posicao_barras1 + largura_barra)
plt.xticks(posicao_barras2, nomes_paises, rotation=90)

# Adicionando legenda
plt.legend()

# Exibir o gráfico
plt.show()
```

No ano de 2018, neste gráfico de barras empilhada, vemos Burundi ainda como o país com menor taxa de Liberdade, mas Generosidade excedendo sua própria taxa de Liberdade de escolha e com valores próximos da Finlândia.

Dados ano 2018

```python
dados2018 = pd.read_csv('/content/2018.csv')
dados2018
```

Apresentando os dados dos 5 melhores colocados e 5 piores com gráfico de barras empilhada.

```python
# Colunas desejadas
nomes_paises = pd.concat([dados2018['Country or region'][:5], dados2018['Country or region'][-5:]])
valores1 = pd.concat([dados2018['Freedom to make life choices'][:5], dados2018['Freedom to make life choices'][-5:]])
valores2 = pd.concat([dados2018['Generosity'][:5], dados2018['Generosity'][-5:]])

# Criando o gráfico de barras empilhadas
plt.bar(nomes_paises, valores1, label='Freedom')
plt.bar(nomes_paises, valores2, bottom=valores1, label='Generosity')

# Personalizando o eixo x com os rótulos adequados
plt.xticks(rotation=90)

# Adicionando legenda
plt.legend()

# Exibir o gráfico
plt.show()
```

Para o ano de 2019, temos um mapa coroplético, que demonstra pela alteração das cores, o grau da percepção de Liberdade de escolha. Encontramos o Afeganistão com o seu valor completamente zerado. Em 2019 houve o ápice de uma crise em decorrência do conflito armado, que desde 2009 afligia esse país. Os talibãs foram responsáveis por 47% das vítimas civis (1.301 mortos e 3.603 feridos) neste ano, o que representa 21% de aumento com relação ao ano de 2018.

Dados ano 2019

```python
dados2019 = pd.read_csv('/content/2019.csv')
dados2019
```

Apresentando os dados do ano de 2019 dos 5 melhores colocados e 5 piores com um mapa coroplético.

```python
# Ordenar os dados pela coluna 'Freedom' em ordem ascendente
data_sorted = dados2019.sort_values('Freedom to make life choices')

# Obter os 5 primeiros e 5 últimos países
primeiros_paises = data_sorted.head(5)
ultimos_paises = data_sorted.tail(5)

# Plotar o mapa coroplético dos 5 primeiros países
fig = px.choropleth(data_frame=primeiros_paises,
                    locations='Country or region',
                    locationmode='country names',
                    color='Freedom to make life choices',
                    hover_name='Country or region',  # Adicionar os nomes dos países como texto de sobreposição
                    title='Top 5 Países - Liberdade')
fig.show()

# Plotar o mapa coroplético dos 5 últimos países
fig = px.choropleth(data_frame=ultimos_paises,
                    locations='Country or region',
                    locationmode='country names',
                    color='Freedom to make life choices',
                    hover_name='Country or region',  # Adicionar os nomes dos países como texto de sobreposição
                    title='5 Países Menos - Liberdade')
fig.show()
```

Chegamos ao ano de 2020. Vemos Afeganistão em último lugar, mas com um grau de percepção de liberdade que é baixa mas não zerada. Observamos também Ruanda que tem uma taxa quase que aproximada com Suíça, que está na terceira colocação dos melhores países.

Dados ano 2020

```python
dados2020 = pd.read_csv('/content/2020.csv')
dados2020
```

Apresentação dos dados ano 2020 dos 5 países melhores colocados e 5 piores com um gráfico de bolhas.

```python
# Ordenar os dados com base nos valores Generosidade
dados_ordenados = dados2020.sort_values('Generosity')

# Selecionar os 5 primeiros e 5 últimos países
primeiros_ultimos = pd.concat([dados_ordenados.head(5),
                               dados_ordenados.tail(5)])

# Extrair as colunas desejadas
nomes_paises = primeiros_ultimos['Country name']
valores1 = primeiros_ultimos['Freedom to make life choices']
valores2 = primeiros_ultimos['Generosity']

# Definir o tamanho das bolhas com base nos valores
tamanho_bolhas = primeiros_ultimos['Freedom to make life choices']

## Criar o gráfico de bolhas
plt.scatter(valores1, valores2, s=tamanho_bolhas*100, alpha=0.7)  # Ajuste o multiplicador para aumentar o tamanho das bolhas

# Adicionar os nomes dos países dentro das bolhas
for i, pais in enumerate(nomes_paises):
    plt.text(valores1.iloc[i], valores2.iloc[i], pais, ha='center', va='center')

# Personalizar o gráfico
plt.xlabel('Freedom to make life choices')
plt.ylabel('Generosity')
plt.title('Gráfico de Bolhas')

# Exibir o gráfico
plt.show()
```

Para 2021, nesse gráfico de linhas, encontramos Generosidade com valores negativos em vários países, inclusive na Finlândia que está na primeira colocação. Já na percepção de Liberdade, temos valores bem aproximados de alguns países que estão na última colocação para os que estão nos 5 primeiros colocados.

Dados ano 2021

```python
dados2021 = pd.read_csv('/content/2021.csv')
dados2021
```

Apresentação dos dados ano 2021 dos 5 países melhores colocados e 5 piores com gráfico de linhas.

```python
# Ordenar os dados com base nos valores de Liberdade
dados_ordenados = dados2021.sort_values('Freedom to make life choices')

# Selecionar os 5 primeiros e 5 últimos países
primeiros_paises = dados_ordenados.head(5)
ultimos_paises = dados_ordenados.tail(5)

# Criar o gráfico de linhas para os 5 primeiros países (azul)
plt.plot(primeiros_paises['Country name'], primeiros_paises['Freedom to make life choices'], 'red', marker='o', label='5 Últimos Liberdade')
plt.plot(primeiros_paises['Country name'], primeiros_paises['Generosity'], 'pink', marker='o', label='5 Últimos Generosidade')

# Criar o gráfico de linhas para os 5 últimos países (vermelho)
plt.plot(ultimos_paises['Country name'], ultimos_paises['Freedom to make life choices'], 'blue', marker='o', label='5 Primeiros Liberdade')
plt.plot(ultimos_paises['Country name'], ultimos_paises['Generosity'], 'green', marker='o', label='5 Primeiros Generosidade')

# Personalizar o gráfico
plt.xlabel('Países')
plt.ylabel('Freedom to make life choices')
plt.title('Gráfico de Linhas')
plt.legend()

# Rotacionar os rótulos dos países para facilitar a leitura
plt.xticks(rotation=45)

# Exibir o gráfico
plt.show()
```

Nesse gráfico de dispersão, observamos novamente que Afeganistão está zerado na sua percepção de Liberdade. Apesar que está bem distante os graus da percepção de liberdade entre os primeiros países e ou últimos, vemos que em relação a generosidade, os valores estão bem aproximados. Em Ruanda, seu valor está se equiparando com Dinamarca, que está em segundo lugar, enquanto que Finlândia, que é a primeira do ranking, está com o seu valor quase igual a Zimbabue.

Dados ano 2022

```python
dados2022 = pd.read_csv('/content/2022.csv')
dados2022
```

Apresentação dos dados ano 2022 dos 5 países melhores colocados e 5 piores colocados com gráfico de dispersão.

```python
# Selecionar os 5 primeiros e 5 últimos países
primeiros_paises = dados_ordenados.head(5)
ultimos_paises = dados_ordenados.tail(5)

# Extrair as colunas desejadas substituindo vírgula por ponto
valores1_primeiros = primeiros_paises['Explained by: Freedom to make life choices'].str.replace(',', '.').astype(float)
valores2_primeiros = primeiros_paises['Explained by: Generosity'].str.replace(',', '.').astype(float)
valores1_ultimos = ultimos_paises['Explained by: Freedom to make life choices'].str.replace(',', '.').astype(float)
valores2_ultimos = ultimos_paises['Explained by: Generosity'].str.replace(',', '.').astype(float)

# Obter os nomes dos países
nomes_primeiros = primeiros_paises['Country']
nomes_ultimos = ultimos_paises['Country']

# Criar o gráfico de dispersão
plt.scatter(valores1_primeiros, valores2_primeiros, color='red', label='Últimos 5')
plt.scatter(valores1_ultimos, valores2_ultimos, color='blue', label='Top 5')

# Adicionar os nomes dos países
for i, nome in enumerate(nomes_primeiros):
    plt.annotate(nome, (valores1_primeiros.iloc[i], valores2_primeiros.iloc[i]), fontsize=8, ha='right', va='bottom')

for i, nome in enumerate(nomes_ultimos):
    plt.annotate(nome, (valores1_ultimos.iloc[i], valores2_ultimos.iloc[i]), fontsize=8, ha='right', va='bottom')

# Personalizar o gráfico
plt.xlabel('Freedom to make life choices')
plt.ylabel('Generosity')
plt.title('Gráfico de Dispersão')

# Adicionar legenda
plt.legend()

# Exibir o gráfico
plt.show()
```
